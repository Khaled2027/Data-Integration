```python
# Austin
airbnb_austin_reviews_df= spark.read.format("csv").options(delimiter=",", header=True,inferSchema=True).load("/Volumes/data_integration/project/datasets/Data/Airbnb_Austin_Reviews.csv")
# Berlin
airbnb_berlin_reviews_df=spark.read.format("csv").options(delimiter=",", header=True,inferSchema=True).load("/Volumes/data_integration/project/datasets/Data/Airbnb_Berlin_Reviews.csv")
# Barcelona
airbnb_barcelona_reviews_df=spark.read.format("csv").options(delimiter=",", header=True,inferSchema=True).load("/Volumes/data_integration/project/datasets/Data/Airbnb_Barcelona_Reviews.csv")
#Bangkok
airbnb_bangkok_reviews_df=spark.read.format("csv").options(delimiter=",", header=True,inferSchema=True).load("/Volumes/data_integration/project/datasets/Data/Airbnb_Bangkok_Reviews.csv")

reviews_dict= {
    "airbnb_austin_reviews_df": airbnb_austin_reviews_df,
    "airbnb_berlin_reviews_df":airbnb_berlin_reviews_df,
    "airbnb_barcelona_reviews_df": airbnb_barcelona_reviews_df,
    "airbnb_bangkok_reviews_df": airbnb_bangkok_reviews_df
}
def display_schema(df):
    for key, value in df.items():
        try:
             print(f"The schema for {key}: ")
             value.printSchema()
        except:
            print(f"failed to load schema for {key}")
display_schema(reviews_dict)
```

```python
reviews_raw_dfs_dict= {
    "airbnb_austin_reviews_df_raw": airbnb_austin_reviews_df,
    "airbnb_berlin_reviews_df_raw":airbnb_berlin_reviews_df,
    "airbnb_barcelona_reviews_df_raw": airbnb_barcelona_reviews_df,
    "airbnb_bangkok_reviews_df_raw": airbnb_bangkok_reviews_df
}
listings_raw_dfs_dict= {
    "airbnb_austin_listings_df_raw": airbnb_austin_listings_df,
    "airbnb_berlin_listings_df_raw": airbnb_berlin_listings_df,
    "airbnb_barcelona_listings_df_raw": airbnb_barcelona_listings_df,
    "airbnb_bangkok_listings_df_raw": airbnb_bangkok_listings_df
}

calender_raw_dfs_dict= {
    "airbnb_austin_calender_df_raw": airbnb_austin_calender_df,
    "airbnb_berlin_calender_df_raw":airbnb_berlin_calender_df,
    "airbnb_barcelona_calender_df_raw": airbnb_barcelona_calender_df,
    "airbnb_bangkok_calender_df_raw": airbnb_bangkok_calender_df
}
neighbourhoods_raw_dfs_dict= {
    "airbnb_austin_neighbourhoods_df_raw": airbnb_austin_neighbourhoods_df,
    "airbnb_berlin_neighbourhoods_df_raw":airbnb_berlin_neighbourhoods_df,
    "airbnb_barcelona_neighbourhoods_df_raw": airbnb_barcelona_neighbourhoods_df,
    "airbnb_bangkok_neighbourhoods_df_raw": airbnb_bangkok_neighbourhoods_df
}

```

```python
def save_to_table(df):
    for key, values in df.items():
        try:
            values.write.mode("overwrite").saveAsTable(key) 
        except:
            print(f"Failed to save {key} as a table")

save_to_table(reviews_raw_dfs_dict)
save_to_table(listings_raw_dfs_dict)
save_to_table(calender_raw_dfs_dict)
save_to_table(neighbourhoods_raw_dfs_dict)
```
